```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('weather')



plt.plot(x,y, color='green', linewidth=5, linestyle='dotted')
```

```python
plt.plot(x,y, 'rD')
```

```python
plt.plot(x,y, color='red', marker='D', linestyle='' )
```

```python
plt.plot(x,y,color='green',marker='+',linestyle='', markersize=20)
```

```python
plt.plot(x,y,color='green', alpha=0.2)
```
